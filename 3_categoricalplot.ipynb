```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
df.sample(2)
```

```python
# Univariate and categorical
plt.figure(figsize=(4,2))
sns.countplot(df, x='day')
plt.title('countplot')
```

```python
# barplot univariate and bivariate
plt.figure(figsize=(4,2))
sns.barplot(df, x='day', hue='time')
```

```python
#we use numerical data for univariate
#for bi variate we use a numerical data and categorical data

plt.figure(figsize=(6,2))
plt.subplot(1,2,2)
sns.boxenplot(df, x='tip')
plt.title('univariate')

plt.subplot(1,2,1)
sns.boxplot(df, x='day', y='total_bill')
plt.title('Multi variate')
```

```python
plt.figure(figsize=(6,2))
plt.subplot(1,2,1)
sns.violinplot(df, x='total_bill',y= 'day')
plt.title('bi / multi variate')

plt.subplot(1,2,2)
sns.violinplot(df, x = 'day')
plt.title('univariate')
```

```python
#bivariate and multi variate 
plt.figure(figsize=(4,3))
sns.swarmplot(df, x='tip')
plt.plot()
```
